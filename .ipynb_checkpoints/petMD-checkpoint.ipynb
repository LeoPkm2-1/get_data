## 1. Get data

## 1.1. Dog Breed

```python
main_petMD_url = 'https://www.petmd.com'
url_for_dog_breed = 'https://www.petmd.com/dog/breeds'
```

```python
from bs4 import BeautifulSoup
headers = {
    "User-Agent":'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36'
}
res = requests.get(url_for_dog_breed, headers=headers)
if res.status_code!=200:
    raise Exception("not get data")
soup = BeautifulSoup(res.content, 'html.parser')
html_content = soup.prettify()
# f = open('foo_file.txt','w')
# f.write(html_content)
# html_content
```

```python
import pandas as pd
breed_names = []
breed_urls = []
link_tags = soup.select('.kib-grid.kib-grid--align-left.kib-grid--valign-top a[href^="/dog/breeds/"]')
# type(link_tags)
# len(link_tags)

for link in link_tags:
    name = link.text
    name = name.strip() if name is not None else ''
    breed_names.append(name)
    sub_domain = link.get("href")
    sub_domain = sub_domain.strip() if sub_domain is not None else ''
    breed_urls.append(main_petMD_url+sub_domain)
# save to file
df = pd.DataFrame({'Breed_name':breed_names,'url':breed_urls})
df.to_csv('link_file/PetMD-Dog_Breeds-Post-Links.csv',index=False)
# show some link
breed_urls[:10]
```

## 1.2. Dog puppies

```python
main_petMD_url = 'https://www.petmd.com'
url_fmt_for_puppies_page = 'https://www.petmd.com/dog/puppy/p/{}'

from bs4 import BeautifulSoup
import pandas as pd
headers = {
    "User-Agent":'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36'
}

post_names = []
post_urls = []

total_page = 17

for pagenum in range(1,total_page+1):
    print('\n\npage: '+str(pagenum))
    res = requests.get(url_fmt_for_puppies_page.format(pagenum), headers=headers)
    
    if res.status_code != 200:
        raise Exception("not get data")
    soup = BeautifulSoup(res.content, 'html.parser')

    post_selector = 'section[data-testid="section-articles"] .article_card_articleCard__UmssU'
    
    # link_tags = soup.select(post_selector +' a[href^="/dog/"]')
    link_tags = soup.select(post_selector)
    post_name_tags = soup.select(post_selector + ' .article_card_articleHeading__E8tYm') 
    for idx,link in enumerate(link_tags):
        sub_domain=link.select('a')[0].get("href")
        sub_domain = sub_domain.strip() if sub_domain is not None else ''
    
        url = main_petMD_url + sub_domain
        post_urls.append(url)

        post_name = post_name_tags[idx].text
        post_names.append(post_name)
        
        print(post_name,url)
# len(post_urls)
# post_names
df = pd.DataFrame({'post_name':post_names,'Url':post_urls})
df.to_csv('link_file/PetMD-Dog_Puppies-Post-Links.csv',index=False)
post_urls[:5]
```

## 1.7. Dog Diseases

```python
main_petMD_url = 'https://www.petmd.com'
url_for_Dog_Diseases = 'https://www.petmd.com/dog/conditions'


from bs4 import BeautifulSoup
import pandas as pd
headers = {
    "User-Agent":'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36'
}
res = requests.get(url_for_Dog_Diseases, headers=headers)
if res.status_code!=200:
    raise Exception("not get data")
soup = BeautifulSoup(res.content, 'html.parser')
html_content = soup.prettify()
# f = open('foo_file.txt','w')
# f.write(html_content)
# html_content


post_names = []
post_urls = []

class_selector = ".kib-grid__item.kib-grid__item--span-4@min-xs.kib-grid__item--span-4@md.kib-grid__item--span-4@min-lg"
link_tags = soup.select('div[class="kib-grid__item kib-grid__item--span-4@min-xs kib-grid__item--span-4@md kib-grid__item--span-4@min-lg"] a[href^="/dog/conditions/"]')
link_tags
# type(link_tags)
len(link_tags)


for link in link_tags:
    name = link.text
    name = name.strip() if name is not None else ''
    post_names.append(name)
    
    sub_domain = link.get("href")
    sub_domain = sub_domain.strip() if sub_domain is not None else ''
    post_urls.append(main_petMD_url+sub_domain)
    
    # print(name,main_petMD_url+sub_domain)
    
# save to file
df = pd.DataFrame({'post_name':post_names,'Url':post_urls})
df.to_csv('link_file/PetMD-Dog_Diseases-Post-Links.csv',index=False)
# show some link
post_urls[:10]
```

## 1.8. Dog Nutrition

```python
main_petMD_url = 'https://www.petmd.com'
url_fmt_for_DogNutrition_page = 'https://www.petmd.com/dog/centers/nutrition/p/{}'
from bs4 import BeautifulSoup
import pandas as pd
headers = {
    "User-Agent":'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/119.0.0.0 Safari/537.36'
}


post_names = []
post_urls = []
for pagenum in range(1,7):
    # print('\n\npage: '+str(pagenum))
    res = requests.get(url_fmt_for_DogNutrition_page.format(pagenum), headers=headers)
    if res.status_code != 200:
        raise Exception("not get data")
    soup = BeautifulSoup(res.content, 'html.parser')

    css_selector_url = 'div[class^="kib-grid__item kib-grid__item--span-4@min-xs kib-grid__item--span-8@md kib-grid__item--span-12@min-lg"] div.article_card_articleCard__UmssU'
    link_tags = soup.select(css_selector_url+' a[href^="/dog/"]')
    post_name_tags = soup.select(css_selector_url + ' div.article_card_articleHeading__E8tYm')
    for idx,link in enumerate(link_tags):
        sub_domain = link.get("href")
        sub_domain = sub_domain.strip() if sub_domain is not None else ''
    
        url = main_petMD_url + sub_domain
        post_urls.append(url)
        
        post_name = post_name_tags[idx].text.strip()
        post_names.append(post_name)
        
        # print(post_name,url)
# len(post_urls)
# post_names
df = pd.DataFrame({'post_name':post_names,'Url':post_urls})
df.to_csv('link_file/PetMD-Dog_Nutrition-Post-Links.csv',index=False)
post_urls[:5]


```
